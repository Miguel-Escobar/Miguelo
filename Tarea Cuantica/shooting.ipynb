```python
# Ejemplo MC 2: harmonic oscillator; casos par e impar
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math

L=10
N=10000
M=100
deltax=L/N
psi=np.empty(N)
phi=np.empty(N)
k2=np.empty(N)
x=np.empty(N)
zero=np.empty(N)
pot=np.empty(N)
emin=.5
emax=7.1
deltae = (emax-emin)/M 

def v(x):
    return x**2
zero[:]=0.
for j in range(M):
    print('j=',j)
    e = emin + j*deltae
    psi[0] = 0
    x[0] = 0
    k2[0] = v(x[0])-e
    phi[0] = 1#k2[0]*psi[0]*deltax/2
    pot[0] = v(x[0])
    for i in range(N-1):
        psi[i+1] = psi[i] + deltax*phi[i]
        phi[i+1] = phi[i] + deltax*psi[i]*k2[i]
        x[i+1] = (i+1)*deltax
        k2[i+1] = v(x[i+1])-e
        pot[i+1] = v(x[i+1])
    plt.figure(figsize=(10,10))
    plt.xlabel('$x$')
    plt.ylabel('$psi$')
    plt.ylim(-2,2)
#    eround = round(e, 4)
    eround = "{0:.4f}".format(e)
    plt.plot(x,psi**2,color='blue', label='e='+str(eround))#'bo', markersize=1)
    plt.plot(x,zero,color='red')
    plt.plot(x,pot,color='green')
    plt.legend(loc='upper right', frameon=False)
#    plt.plot(xs,v_theo,color='red',linestyle='dashed')
    plt.tight_layout()
    if j < 10:
        plt.savefig('shooting00'+str(j), dpi=None)
#if j < 10:
#        plt.savefig('test00'+str(j), dpi=None)
    else:
        if j < 100:
            plt.savefig('shooting0'+str(j), dpi=None)
        else:
            plt.savefig('shooting'+str(j), dpi=None)
    plt.close()
```

```python
# Ejemplo MC 4: hydrogen atom (con leapfrog)
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math

L=30
N=10000
M=100
#R=6
deltax=L/N
psi=np.empty(N)
phi=np.empty(N)
k2=np.empty(N)
x=np.empty(N)
zero=np.empty(N)
pot=np.empty(N)
emin=-1.1
emax=-0.06
deltae = (emax-emin)/M 

l=0
def v(x,l):
    return - 2*x**-1 + l*(l+1)*x**-2

zero[:]=0.
psiold=0
for j in range(M):
    print('j=',j)
    e = emin + j*deltae
    
    psi[0] = 0.01
    x[0] = 30.
    k2[0] = v(x[0],l)-e
    phi[0] = -np.sqrt(k2[0])*psi[0] - k2[0]*psi[0]*deltax/2 #k2[0]*psi[0]*deltax/2
    pot[0] = v(x[0],l)
    for i in range(N-1):
        psi[i+1] = psi[i] - deltax*phi[i]
        phi[i+1] = phi[i] - deltax*psi[i]*k2[i]
        x[i+1] = x[0]-(i+1)*deltax
        k2[i+1] = v(x[i+1],l)-e
# bisección para obtener las autoenergías
    psinew=psi[N-1]-deltax*phi[N-1]
    if psinew*psiold < 0:
        eg = e
        ec = e - deltae
        deltai = deltae
        while abs(eg - ec) > deltae*.01:
            deltai = deltai/2
            ei = ec + deltai
            psi[0] = 0.01
            x[0] = 30.
            k2[0] = v(x[0],l)-ei
            phi[0] = -np.sqrt(k2[0])*psi[0] - k2[0]*psi[0]*deltax/2 #k2[0]*chi[0]*deltax/2
            for i in range(N-1):
                psi[i+1] = psi[i] - deltax*phi[i]
                phi[i+1] = phi[i] - deltax*psi[i]*k2[i]
                x[i+1] = x[0]-(i+1)*deltax
                k2[i+1] = v(x[i+1],l)-ei
            psinewi=psi[N-1]-deltax*phi[N-1]
            if psinewi*psiold < 0:
                eg = ei
            else:
                ec = ei
        print('autoenergía=',(ec+eg)*.5)
    psiold=psinew
# fin de bisección
    plt.figure(figsize=(10,10))
    plt.xlabel('$x$')
    plt.ylabel('$psi$')
#    plt.ylim(-2,2)
    eround = "{0:.4f}".format(e)
    plt.plot(x,psi,color='blue', label='e='+str(eround))#'bo', markersize=1)
    plt.plot(x,zero,color='red')
    plt.legend(loc='upper right', frameon=False)
    plt.tight_layout()
    if j < 10:
        plt.savefig('shooting00'+str(j), dpi=None)
    else:
        if j < 100:
            plt.savefig('shooting0'+str(j), dpi=None)
        else:
            plt.savefig('shooting'+str(j), dpi=None)
    plt.close()
```

```python
# Ejemplo MC 5: hydrogen atom (con RK4)
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math

L=30
N=100
M=100
deltax=L/N
dx=0.001
psi=np.empty(N+1)
phi=np.empty(N+1)
k2a=np.empty(N+1)
k2b=np.empty(N+1)
k2c=np.empty(N+1)
x=np.empty(N+1)
zero=np.empty(N+1)
#pot=np.empty(N)
emin=-1.1
emax=-0.06
deltae = (emax-emin)/M 

l=0
def v(x,l):
    return - 2*x**-1 + l*(l+1)*x**-2

zero[:]=0.
psiold=0
for j in range(M):
    print('j=',j)
    e = emin + j*deltae
    
    psi[0] = 0.01
    x[0] = L
    k2[0] = v(x[0],l)-e
    phi[0] = -np.sqrt(k2[0])*psi[0]# - k2[0]*psi[0]*deltax/2 #k2[0]*psi[0]*deltax/2
#    pot[0] = v(x[0],l)
#    deltax=L/N
    for i in range(N):
        if i == N-1:
            deltax=(1-dx)*deltax
        k2a[i] = v(x[i],l)-e
        k1_psi = deltax*phi[i]
        k1_phi = deltax*psi[i]*k2a[i]
        k2b[i] = v(x[i]-.5*deltax,l)-e
        k2_psi = deltax*(phi[i]-.5*k1_phi)
        k2_phi = deltax*(psi[i]-.5*k1_psi)*k2b[i]
        k3_psi = deltax*(phi[i]-.5*k2_phi)
        k3_phi = deltax*(psi[i]-.5*k2_psi)*k2b[i]
        k2c[i] = v(x[i]-deltax,l)-e
        k4_psi = deltax*(phi[i]-k3_phi)
        k4_phi = deltax*(psi[i]-k3_psi)*k2c[i]
        psi[i+1] = psi[i] - (k1_psi+2*k2_psi+2*k3_psi+k4_psi)/6.
        phi[i+1] = phi[i] - (k1_phi+2*k2_phi+2*k3_phi+k4_phi)/6.
        x[i+1] = x[i]-deltax
    deltax=L/N
    psi[N] = psi[N] - dx*deltax*phi[N]    
# bisección para obtener las autoenergías
    psinew=psi[N]#-deltax*phi[N-1]
    if psinew*psiold < 0:
        eu = e
        ed = e - deltae
        deltai = deltae
        while abs(eu - ed) > deltae*.01:
            deltai = deltai/2
            ei = ed + deltai
            psi[0] = 0.01
            x[0] = L
            k2[0] = v(x[0],l)-e
            phi[0] = -np.sqrt(k2[0])*psi[0]# - k2[0]*psi[0]*deltax/2 #k2[0]*psi[0]*deltax/2
            for i in range(N):
                if i == N-1:
                    deltax=(1-dx)*deltax
                k2a[i] = v(x[i],l)-e
                k1_psi = deltax*phi[i]
                k1_phi = deltax*psi[i]*k2a[i]
                k2b[i] = v(x[i]-.5*deltax,l)-e
                k2_psi = deltax*(phi[i]-.5*k1_phi)
                k2_phi = deltax*(psi[i]-.5*k1_psi)*k2b[i]
                k3_psi = deltax*(phi[i]-.5*k2_phi)
                k3_phi = deltax*(psi[i]-.5*k2_psi)*k2b[i]
                k2c[i] = v(x[i]-deltax,l)-e
                k4_psi = deltax*(phi[i]-k3_phi)
                k4_phi = deltax*(psi[i]-k3_psi)*k2c[i]
                psi[i+1] = psi[i] - (k1_psi+2*k2_psi+2*k3_psi+k4_psi)/6.
                phi[i+1] = phi[i] - (k1_phi+2*k2_phi+2*k3_phi+k4_phi)/6.
                x[i+1] = x[i]-deltax
            deltax=L/N
            psi[N] = psi[N] - dx*deltax*phi[N]    
            psinewi=psi[N]
            if psinewi*psiold < 0:
                eu = ei
            else:
                ed = ei
        print('autoenergía=',(eu+ed)*.5)
    psiold=psinew
# fin de bisección
    plt.figure(figsize=(10,10))
    plt.xlabel('$x$')
    plt.ylabel('$psi$')
#    plt.ylim(-2,2)
    eround = "{0:.4f}".format(e)
    plt.plot(x,psi,color='blue', label='e='+str(eround))#'bo', markersize=1)
    plt.plot(x,zero,color='red')
    plt.legend(loc='upper right', frameon=False)
    plt.tight_layout()
    if j < 10:
        plt.savefig('shooting00'+str(j), dpi=None)
    else:
        if j < 100:
            plt.savefig('shooting0'+str(j), dpi=None)
        else:
            plt.savefig('shooting'+str(j), dpi=None)
    plt.close()
```
